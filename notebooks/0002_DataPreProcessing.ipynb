```python
os.chdir(r'../data/allYaws/')
all_files=glob.glob('*parquet')
print(all_files)
```

# Importing all data into dataframe

```python
posYaw_df=pd.read_parquet('allNegYaws.parquet')
```

```python
negYaw_df=pd.read_parquet('allPostiveYaws.parquet')
```

```python
posYaw_df.head()
```

```python
negYaw_df.head()
```

# Concate all the data

```python
allNegPos_df = pd.concat([posYaw_df, negYaw_df])
allNegPos_df = allNegPos_df.reset_index(drop=True)
```

```python
allNegPos_df.head()
```

# Dividing dataframe into targets_df and inputs_df

```python
target_cols = ['wind_speed', 'yaw_error', 'turbulence']
targets_df = allNegPos_df[target_cols]
inputs_df = allNegPos_df.drop(columns=target_cols)

print('Targets shape:', targets_df.shape)
print('Inputs shape:', inputs_df.shape)
```
